```python
import os
import cv2
import pandas as pd

# Define the path to the dataset
dataset_path = "/content/drive/MyDrive/INDIVIDUALES/ISMA/KOKAS/FRUTAS/train"

# Initialize a list to store metadata
metadata = []

# Iterate over each label folder in the dataset
for label in os.listdir(dataset_path):
    label_path = os.path.join(dataset_path, label)
    print(label)
    # Skip if not a directory
    if not os.path.isdir(label_path):
        continue

    # Iterate over each image file in the label folder
    for image_file in os.listdir(label_path):
        image_path = os.path.join(label_path, image_file)
        # Load the image with OpenCV
        image = cv2.imread(image_path)

        # Skip if the image couldn't be loaded
        if image is None:
            continue

        # Get image dimensions
        height, width, channels = image.shape

        # Append metadata for the image
        metadata.append({
            "label": label,
            "file_path": image_path,
            "height": height,
            "width": width,
            "channels": channels
        })

# Convert metadata to a DataFrame
metadata_df = pd.DataFrame(metadata)
print(metadata_df)
# Save to CSV
metadata_df.to_csv("metadata.csv", index=False)


print("Metadata saved successfully!")
```
